```python
#Remove duplicate rows using .drop_duplicates() or Excel’s “Remove Duplicates”.
df.drop_duplicates()
```

```python
#Standardize text values like gender, country names, etc.

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Option: Drop rows with any missing values
df = df.dropna()
```

```python
df
```

```python
#Standardize Text Values (e.g., Gender, country)
# Clean gender column if it exists
if 'gender' in df.columns:
    df['gender'] = df['gender'].str.strip().str.lower().replace({
        'f': 'female', 'm': 'male', 'fem': 'female', 'mal': 'male'
    })
```

```python
# to find duplicates
print(df.duplicated().sum())
```
